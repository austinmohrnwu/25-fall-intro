## Mean ##

```python
values = make_array(2, 3, 3, 9)
values
```

The mean (or average) of an array is defined to be the sum of all the values in the array
divided by the size of the array.

```python
sum(values)/len(values)
```

```python
# The calculation is built into the np package.
values_mean = np.mean(values)
values_mean
```

```python
# If every item in the array was replaced by the mean, the sum would stay the same.
values2 = make_array(4.25, 4.25, 4.25, 4.25)
sum(values2)
```

# Mean as Center of Gravity

```python
values_table = Table().with_columns('value', values)
values_table
bins_for_display = np.arange(0.5, 10.6, 1)
values_table.hist('value', bins = bins_for_display)
plots.ylim(-0.04, 0.5)
plots.plot([0, 10], [0, 0], color='grey', lw=2)
plots.scatter(4.25, -0.015, marker='^', color='red', s=100)
plots.title('Mean as a Center of Gravity');
```

The average is located at the point that would balance the data if it was a lever with masses.

## Standard Deviation ##

Standard deviation is a measure of how spread out the data is.

Small standard deviation means the data is tightly clustered around the mean.

The larger the standard deviation, the less tight the clustering.

# Computing the Standard Deviation

Step 1. For each value in the array, compute the "mean minus value".

Step 2. Square the results found in Step 1.

Step 3. Average the squared differences from Step 2.

Step 4. Compute the square root of the average in Step 3.

```python
values_table
```

Step 1: Attach a column that shows "mean minus value".

```python
diffs = values_mean - values
values_table = values_table.with_column('diffs', diffs)
values_table
```

Step 2: Attach a column that shows the squared difference for each row.

```python
square_diffs = diffs ** 2
values_table = values_table.with_column('square diffs', square_diffs)
values_table
```

Step 3: Compute the average of the squared difference column. (No need to attach to table.)

```python
average_square_diffs = np.mean(square_diffs)
average_square_diffs
```

Step 4: Compute the square root of the sum from Step 3.

```python
std_dev = average_square_diffs ** 0.5
std_dev
```

Or just use the built-in function.

```python
np.std(values)
```
